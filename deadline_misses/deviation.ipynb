```python
import numpy as np
import pandas as pd

nominal = pd.read_csv("./p2_1hit0miss.tsv", sep="\t")
two_miss = pd.read_csv("./p2_1hit2miss.tsv", sep="\t")

dy = nominal['y'] - two_miss['y']
print(max(dy), np.argmax(dy))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.legend(['nominal', '2 miss'])
plt.plot(nominal['y'])
plt.plot(two_miss['y'])
```

```python

plt.plot(np.abs(two_miss['y'] - nominal['y']))
plt.legend(['deviation'])
```
